```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
dataets = pd.read_csv("online_retail.csv")
```

```python
dataets.head()
```

```python
#Preprocessing data (menghilangkan spasi deskripsi) serta menghapus baris yang tidak memiliki invoice number

dataets['Description'] = dataets['Description'].str.strip()
dataets.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
  
dataets['InvoiceNo'] = dataets['InvoiceNo'].astype('str') 
dataets = dataets[~dataets['InvoiceNo'].str.contains('C')]
```

```python
basket = dataets.groupby('InvoiceNo')
invoice = []
for i,j in basket:
    invoice.append(list(j['Description'].map(str)))
```

```python
data = pd.DataFrame(invoice)
```

```python
data
```

```python
records=[]
for i in range (0,22064):
    records.append([str(data.values[i,j]) for j in range(0,1114) if str(data.loc[i,j])!="None"])
```

```python
association_rules = apriori(records, min_support=0.01,
                           min_confidence=0.8,
                           min_lift=1.3, min_length=2)
association_result=list(association_rules)
```

```python
for item in association_result:
#firts index of the inner list
#contains base item and add item
    pair = item[0]
    items = [a for a in pair]
    print("rule: " + items [0] + " ->" + items[1])

    #second index of the inner list
    print("support: " + str(item[1]))

    #third index of the list located at 10th
    #of the third index of the inner list

    print("confidence: " + str(item[2][0][2]))
    print("lift: " + str(item[2][0][3]))
    print("=======================================")
```

#  Kesimpulan

Dari hasil diatas, rule: HERB MARKER ROSEMARY ->HERB MARKER THYME memiliki persentase lift tertinggi sebesar 85,73%. Artinya, perusahaan disarankan untuk meletakkan kedua produk tersebut berdekatan, atau dalam kasus retail online kedua barang tersebut sebaiknya diletakkan di etalase yang sama atau saling merekomendasikan barang tersebut satu sama lain saat konsumen sedang melihat salah satu barang tersebut.
